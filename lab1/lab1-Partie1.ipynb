```python
#Azaiez Mohamed amine & Jlassi Olfa

# imread and display image
import cv2 



# path 
path = r'C:\Users\asus\Desktop\opencv\s1\bird.jpg'
  
# Reading an image in default mode
image = cv2.imread(path)
  
# Window name in which image is displayed
window_name = 'image'
  
# Using cv2.imshow() method 
# Displaying the image 
cv2.imshow(window_name, image)
  
#waits for user to press any key 
#(this is necessary to avoid Python kernel form crashing)
cv2.waitKey(0) 
  
#closing all open windows 
cv2.destroyAllWindows() 

def image_augmentation(image):
    
    #flipping
    image2 = cv2.flip(image, 0)

    # Displaying the image
    cv2.imshow("flipping", image2)
    cv2.waitKey(0)
    
    #Cropping
    crop = image[50:180, 100:300]  
    cv2.imshow('cropped', crop)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    #Scaling
    bigger = cv2.resize(image, (1050, 1610))
    cv2.imshow('Scaling', bigger)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    #Rotating

    # Extracting height and width from 
    # image shape
    height, width = image.shape[:2]

    # get the center coordinates of the
    # image to create the 2D rotation
    # matrix
    center = (width/2, height/2)

    rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=90, scale=1)

    # rotate the image using cv2.warpAffine 
    # 90 degree anticlockwise
    rotated_image = cv2.warpAffine(src=image, M=rotate_matrix, dsize=(width, height))

    cv2.imshow("rotated image:", rotated_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    #Brightness
    import numpy as np

    matrix = np.ones(image.shape, dtype = "uint8") * 120 
    image2 = cv2.add(image, matrix)
    cv2.imshow("Brightness:", image2)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Median Blur
    median = cv2.medianBlur(image, 5)
    cv2.imshow('Median Blurring', median)
    cv2.waitKey(0)

    #Sharpening
    Kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    sharpened = cv2.filter2D(image, -1, Kernel)
    cv2.imshow('sharpened', sharpened)
    cv2.waitKey(0)
    
image_augmentation(image)
```
